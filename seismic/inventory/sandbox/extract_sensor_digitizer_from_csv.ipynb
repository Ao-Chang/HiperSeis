```python
import pandas as pd
```

```python
# where is the input CSV file
csvfile ="../../../tests/testdata/FieldSiteVisitLive.csv"

pdf = pd.read_csv(csvfile,delimiter=',')
```

```python
pdf.shape
```

```python
pdf.head()
```

```python
my_columns=["submissionid",
            "submissiondatetime",
            "deviceid",
            "username",
            "Select_Site_ID",
            "GPS_Locations_Subform.GPS_Location_Group.Select_Location_Type",
            "GPS_Locations_Subform.GPS_Location_Group.Set_Location",
            "Replace_Equipment_Yes_No",
            "Asset_Tracking_Subform.Asset_Barcodes_Group.DB_Asset_Type",
            "Asset_Tracking_Subform.Asset_Barcodes_Group.DB_Asset_UniqueID"]
```

```python
mypdf = pdf[my_columns]
```

```python
mypdf.head()
```

```python
my_criteria = ((mypdf['Asset_Tracking_Subform.Asset_Barcodes_Group.DB_Asset_Type'] == "Nanometrics Trillium Compact 120s") | 
(mypdf['Asset_Tracking_Subform.Asset_Barcodes_Group.DB_Asset_Type'] == "Guralp Minimus"))

```

```python
my_criteria
```

```python
my_selected_pdf = mypdf.loc[my_criteria].copy()  # use .copy() to fix "SettingWithCopyWarning"
```

```python
my_selected_pdf.head()
```

```python
my_selected_pdf.shape
```

```python
output_cols = ["submissionid",
            #"deviceid", # iPad_7F8BBE26-0768-46FB-A839-E499270B9A9C
            "Select_Site_ID",
            "Replace_Equipment_Yes_No",
            "Asset_Tracking_Subform.Asset_Barcodes_Group.DB_Asset_Type",
            "Asset_Tracking_Subform.Asset_Barcodes_Group.DB_Asset_UniqueID"]
```

```python
my_selected_pdf[output_cols].to_csv("OA_sensors_digitizers.csv", index=False)
```
